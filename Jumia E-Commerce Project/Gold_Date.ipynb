# Creating Gold_Date table


```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Creating Table structure
```

```python
%%sql
create table if not exists ecom_proj_LH.gold_date
(
    BusinessDate Date,
    Business_Year int,
    Business_Month int,
    Business_Quarter int,
    Business_Week int
)
using DELTA
PARTITIONED by (Business_Year)
```

```python
# Now inser the data into gold date table
# one way we can use the order_date/shipping_date from bronze_sales table
# 2nd way is we can specify the start date range and end date range, -- I'm going to use 2nd menthos
```

```python
start_date = '2015-01-01'
end_date = '2030-12-31'

date_diff = spark.sql("select date_diff('{}','{}')".format(end_date,start_date)).first()[0]
```

```python
date_diff
```

```python
#now generate the range of numebrs from 0 to 5843
ID_Data = spark.range(0,date_diff)
```

```python
date_data = ID_Data.selectExpr("date_add('{}',cast(id as int)) as BusinessDate".format(start_date))
```

```python
# Creating a view called ViewDate
date_data.createOrReplaceTempView("ViewDate")
```

```python
#Now generate columns from viewdate and insert the data into gold_date table
```

```python
%%sql

delete from ecom_proj_LH.gold_date;
```

```python
%%sql

insert into ecom_proj_LH.gold_date
select *,
Year(BusinessDate) as Business_Year,
Month(BusinessDate) as Business_Month,
Quarter(BusinessDate) as Business_Quarter,
weekday(BusinessDate) as Business_Week
from ViewDate
```
