# Creating Gold_OrderReturns dimensions

```python
from pyspark.sql import * 
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Create returns dim table (Gold)
```

```python
%%sql
create table if not exists ecom_proj_LH.Gold_OrderReturn
(
    OrderID string,
    Return string,
    Order_Year int,
    Order_Month int,
    Created_TS TIMESTAMP,
    Modified_TS TIMESTAMP
)
using DELTA
partitioned by(Order_Year,Order_Month)
```

```python
# Now implement incremental  load
```

```python
Max_Date = spark.sql("Select coalesce(max('Modified_TS'),'1900-01-01') from ecom_proj_LH.Gold_OrderReturn").first()[0]
```

```python
Max_Date
```

```python
df = spark.sql("""select 
Order_ID,
Return ,
Order_Year ,
Order_Month ,
Created_TS ,
Modified_TS 
from ecom_proj_LH.bronze_sales where Return = 'Yes' and Modified_TS > '{}'""".format(Max_Date))
```

```python
df.createOrReplaceTempView("ViewReturn")
```

```python
# Now inser the return data in goldreturns table
```

```python
%%sql
insert into ecom_proj_LH.gold_orderreturn
select * from ViewReturn
```
